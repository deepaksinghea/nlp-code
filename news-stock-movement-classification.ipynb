```python
import pandas as pd
```

```python
df = pd.read_csv('news_headline.csv',encoding='ISO-8859-1')
```

```python
df.head(1)
```

```python
df["Date"]= pd.to_datetime(df["Date"])
```

```python
df = df.sort_values(by=['Date'])
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.reset_index()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
X = df.iloc[:,2:27]

y = df.iloc[:,1]
```

```python
X.replace("[^a-zA-Z]"," ",regex = True, inplace = True)
```

```python
l = [i for i in range(25)]
new_index = [str(i) for i in l]
X.columns = new_index
```

```python
for index in new_index:
    X[index] = X[index].str.lower()
    X[index] = [lemmatizer.lemmatize(word) for word in X[index] if not word in stopwords.words('english')]
```

```python
headlines = []
for row in range(0,len(X.index)):
    headlines.append(' '.join(str(x) for x in X.iloc[row,0:25]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(headlines, y, test_size = 0.30, random_state=0, shuffle=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```
